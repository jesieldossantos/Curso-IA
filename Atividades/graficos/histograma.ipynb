# O que é histograma?

Em Python, um histograma é uma representação gráfica que mostra a distribuição de um conjunto de dados numéricos. Ele divide os dados em "bins" (intervalos) e conta quantas observações caem em cada bin, exibindo essa contagem como a altura de uma barra. Essencialmente, um histograma ajuda a visualizar a frequência com que diferentes valores ocorrem em um conjunto de dados, permitindo identificar padrões e tendências. 
Como criar um histograma em Python:
1. Importar bibliotecas:
Utilize a biblioteca matplotlib.pyplot para criar o gráfico. 
2. Preparar os dados:
Tenha seus dados numéricos prontos em uma estrutura de dados como uma lista ou um array. 
3. Criar o histograma:
Utilize a função plt.hist() da biblioteca matplotlib.pyplot, passando os dados e, opcionalmente, o número de bins desejado. 
4. Personalizar o histograma:
Você pode ajustar cores, títulos, rótulos dos eixos, etc., para tornar o gráfico mais informativo e visualmente atraente. 

# O que um histograma mostra:
-O que um histograma mostra:
Distribuição dos dados:
Permite ver como os dados estão distribuídos, se são simétricos, enviesados, se possuem picos, etc.

## Frequência:
Indica quantas vezes cada valor (ou intervalo de valores) aparece nos dados. 

### Padrões e tendências:
Ajuda a identificar padrões, como concentrações de dados em certas faixas ou a presença de outliers (valores atípicos). 

-O histograma é uma ferramenta essencial para a análise exploratória de dados em Python, auxiliando na compreensão da distribuição e padrões dos dados numéricos. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
np.random.seed(0)
dados = np.random.normal(0, 1, 1000)
```

```python
sns.set(style='whitegrid')

```

```python
plt.figure(figsize=(8, 6))
sns.histplot(dados, bins=30, kde=True, color='blue')
plt.title('Histograma de Dados Aleatórios')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()




```
