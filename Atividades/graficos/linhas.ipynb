# LINHAS

- um gráfico de linhas (ou gráfico de linhas) é uma visualização de dados que exibe informações como uma série de pontos de dados conectados por segmentos de linha reta. É comumente usado para representar a evolução de uma ou mais variáveis numéricas ao longo do tempo, sendo especialmente útil para analisar tendências e padrões em séries temporais. Bibliotecas como Matplotlib e Seaborn são frequentemente usadas para criar gráficos de linhas em Python. 


# Como funciona o gráfico de linhas:

1. Eixos:
- Um gráfico de linhas possui dois eixos: o eixo x (horizontal), geralmente usado para representar o tempo ou outra variável independente, e o eixo y (vertical), que mostra a variável dependente ou os valores que estão sendo medidos.

2. Pontos de dados:
- Os dados são plotados como pontos no gráfico, e esses pontos são conectados por linhas para mostrar a progressão dos dados.

3. Linhas:
- As linhas conectando os pontos podem ser retas ou curvas, dependendo dos dados e das configurações de visualização.

# Aplicações comuns:

1. Séries temporais:
- O gráfico de linhas é ideal para visualizar dados que mudam ao longo do tempo, como preços de ações, temperatura, vendas de produtos, etc. 

2. Análise de tendências:
Permite identificar tendências ascendentes, descendentes ou estáveis em um conjunto de dados. 

3. Comparação de variáveis:
- É possível plotar várias linhas em um único gráfico para comparar a evolução de diferentes variáveis. 

# Bibliotecas em Python:

1. Matplotlib:
- Uma biblioteca fundamental para visualização de dados em Python, que oferece uma ampla gama de opções para criar e personalizar gráficos de linhas.

2. Seaborn:
- Uma biblioteca construída sobre o Matplotlib, projetada para gráficos estatísticos mais avançados e com aparência mais atraente. 

3. Plotly:
- Uma biblioteca interativa para gráficos, que permite criar gráficos de linhas com zoom, panorâmica e outras funcionalidades interativas.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
# Gerar dados aleatórios
np.random.seed(0)
x = np.arange(1, 11)
y1 = np.random.randint(10, 50, size=10)
y2 = np.random.randint(20, 60, size=10)

```

```python
# Configurar o estilo do Seaborn
sns.set(style='whitegrid')

# Criar o gráfico de linhas
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='Linha 1', marker='o')
plt.plot(x, y2, label='Linha 2', marker='s')

# Adicionar título e rótulos
plt.title('Gráfico de Linhas')
plt.xlabel('X')
plt.ylabel('Y')

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```
