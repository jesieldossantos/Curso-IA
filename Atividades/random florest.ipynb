# RANDOM FLOREST

- é um metaestimador que ajusta uma série de classificadores de árvores de decisão em várias subamostras do conjunto de dados e usa a média para melhorar a precisão preditiva e controlar o sobreajuste.
-  usam a melhor estratégia de divisão, ou seja, equivalente a passar splitter="best"para a subamostra subjacente DecisionTreeClassifier
- O tamanho da subamostra é controlado com o max_samplesparâmetro if bootstrap=True(padrão); caso contrário, todo o conjunto de dados é usado para construir cada árvore.
- A floresta aleatória utiliza uma técnica chamada "bagging" para construir árvores de decisão completas em paralelo a partir de amostras bootstrap aleatórias do conjunto de dados e recursos. Enquanto as árvores de decisão são baseadas em um conjunto fixo de recursos e, frequentemente, superajustadas, a aleatoriedade é crucial para o sucesso da floresta.

![random florest 3.png](<attachment:random florest 3.png>)

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Dados
X = np.array([[12, 1200, 700], [12, 1200, 700], [12, 1200, 700], [12, 1400, 640], [10, 1600, 620]])
y = np.array([3.0, 3.0, 3.0, 3.0, 3.5])

```

```python
# Treinar o modelo
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

```

```python
# Fazer previsões
y_pred = model.predict(X)

```

```python
# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(y, label="Valor real")
plt.plot(y_pred, label="Previsão")
plt.legend()
plt.xlabel("Amostra")
plt.ylabel("Valor")
plt.title("Previsões do modelo")
plt.show()
```
