# Tabela 05

Na analise vemos que a media por genero esta bem proximo de pretensoes salariais

```python
media_salario_genero = df_candidatos_completo.groupby('genero')['pretensoes_salariais'].mean().reset_index()

total = media_salario_genero['pretensoes_salariais'].sum()
media_salario_genero['Porcentagem'] = (media_salario_genero['pretensoes_salariais'] / total) * 100

plt.figure(figsize=(8, 6))
plt.bar(media_salario_genero['genero'], media_salario_genero['Porcentagem'])
plt.title('Porcentagem do Salário por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Porcentagem')
for i, row in media_salario_genero.iterrows():
    plt.text(row['genero'], row['Porcentagem'], f"{row['Porcentagem']:.2f}%", ha='center', va='bottom')
plt.show()

print(media_salario_genero)
```
