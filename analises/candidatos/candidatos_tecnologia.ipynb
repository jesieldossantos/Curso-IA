Aqui vemos que na tabela temos 33.9% sao masculinos, 32,9% e feminino e 33,2% e outros

```python
# Define the mapping for the 'formacao' column
formacao_mapping = {
	'Ensino Médio': 1,
	'Tecnólogo': 2,
	'Graduação': 3,
	'Mestrado': 4,
	'Doutorado': 5
}

porcentagem_genero = df_candidatos_completo['genero'].value_counts(normalize=True) * 100
df_candidatos_completo['porcentagem_genero'] = df_candidatos_completo['genero'].map(porcentagem_genero)
df_candidatos_completo['formacao_numerico'] = df_candidatos_completo['formacao'].map(formacao_mapping)
df_candidatos_completo['porcentagem'] = (df_candidatos_completo['formacao_numerico'] / df_candidatos_completo['formacao_numerico'].sum()) * 100

plt.figure(figsize=(8, 6))
plt.pie(frequencia_genero, labels=frequencia_genero.index, autopct='%1.1f%%')
plt.title('Frequência de Gênero')
plt.show()

print(porcentagem_genero)

```

# Tabela 05

Na analise vemos que a media por genero esta bem proximo de pretensoes salariais

```python
media_salario_genero = df_candidatos_completo.groupby('genero')['pretensoes_salariais'].mean().reset_index()

total = media_salario_genero['pretensoes_salariais'].sum()
media_salario_genero['Porcentagem'] = (media_salario_genero['pretensoes_salariais'] / total) * 100

plt.figure(figsize=(8, 6))
plt.bar(media_salario_genero['genero'], media_salario_genero['Porcentagem'])
plt.title('Porcentagem do Salário por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Porcentagem')
for i, row in media_salario_genero.iterrows():
    plt.text(row['genero'], row['Porcentagem'], f"{row['Porcentagem']:.2f}%", ha='center', va='bottom')
plt.show()

print(media_salario_genero)
```

# Tabela 7

a tabela mostra que do total de pessoas masculinas com algum tipo de deficiencia sao de 42, femininas sao 48 e outros generos e de 33 pesssoas com deficiencia

```python
import numpy as np

pessoaspcd = df_candidatos_completo.groupby('genero')['é_PCD'].apply(lambda x: (x == 'Sim').sum())
frequencia_genero = df_candidatos_completo['genero'].value_counts()
pessoaspcd = df_candidatos_completo.groupby('genero')['é_PCD'].apply(lambda x: (x == 'Sim').sum()).reindex(frequencia_genero.index, fill_value=0)

plt.figure(figsize=(8, 6))
bar_width = 0.4
x = np.arange(len(frequencia_genero))

plt.bar(x - bar_width/2, frequencia_genero.values, width=bar_width, label='Total por Gênero')
for i, v in enumerate(frequencia_genero.values):
    porcentagem = (v / frequencia_genero.sum()) * 100
    plt.text(i - bar_width/2, v + 5, f"{v} ({porcentagem:.1f}%)", ha='center', size=8)

plt.bar(x + bar_width/2, pessoaspcd.values, width=bar_width, label='PCD por Gênero', color='orange')
for i, v in enumerate(pessoaspcd.values):
    if pessoaspcd.sum() > 0:
        porcentagem = (v / pessoaspcd.sum()) * 100
        plt.text(i + bar_width/2, v + 5, f"{v} ({porcentagem:.1f}%)", ha='center', size=8)

plt.xticks(x, frequencia_genero.index, rotation=45)
plt.title('Frequência de Gênero e PCD no Universo de 2700')
plt.xlabel('Gênero')
plt.ylabel('Frequência')
plt.legend()
plt.tight_layout()
plt.show()

print("Frequência de Gênero:")
print(frequencia_genero)
print("\nFrequência de PCD por Gênero:")
print(pessoaspcd)

```
