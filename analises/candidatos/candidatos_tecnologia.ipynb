# Tabela 07

a tabela mostra que do total de pessoas masculinas com algum tipo de deficiencia sao de 42, femininas sao 48 e outros generos e de 33 pesssoas com deficiencia

```python
pessoaspcd = df_candidatos_completo.groupby('genero')['é_PCD'].apply(lambda x: (x == 'Sim').sum())
frequencia_genero = df_candidatos_completo['genero'].value_counts()
pessoaspcd = df_candidatos_completo.groupby('genero')['é_PCD'].apply(lambda x: (x == 'Sim').sum()).reindex(frequencia_genero.index, fill_value=0)

plt.figure(figsize=(8, 6))
bar_width = 0.4
x = np.arange(len(frequencia_genero))

plt.bar(x - bar_width/2, frequencia_genero.values, width=bar_width, label='Total por Gênero')
for i, v in enumerate(frequencia_genero.values):
    porcentagem = (v / frequencia_genero.sum()) * 100
    plt.text(i - bar_width/2, v + 5, f"{v} ({porcentagem:.1f}%)", ha='center', size=8)

plt.bar(x + bar_width/2, pessoaspcd.values, width=bar_width, label='PCD por Gênero', color='orange')
for i, v in enumerate(pessoaspcd.values):
    if frequencia_genero.values[i] > 0:
        porcentagem = (v / frequencia_genero.values[i]) * 100
        plt.text(i + bar_width/2, v + 5, f"{v} ({porcentagem:.1f}%)", ha='center', size=8)

plt.xticks(x, frequencia_genero.index, rotation=45)
plt.title('Frequência de Gênero e PCD no Universo de 2700')
plt.xlabel('Gênero')
plt.ylabel('Frequência')
plt.legend()
plt.tight_layout()
plt.show()

print("Frequência de Gênero:")
print(frequencia_genero)
print("\nFrequência de PCD por Gênero:")
print(pessoaspcd)

```

# analise 09

Com os dados da analise anterior, aqui vemos uma media da idade por regiao

```python
regioes = {
    'Norte': ['AC', 'AM', 'AP', 'PA', 'RO', 'RR', 'TO'],
    'Nordeste': ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'],
    'Sudeste': ['ES', 'MG', 'RJ', 'SP'],
    'Sul': ['PR', 'RS', 'SC'],
    'Centro-Oeste': ['DF', 'GO', 'MS', 'MT']
}

estado_regiao = {}
for regiao, estados in regioes.items():
    for estado in estados:
        estado_regiao[estado] = regiao

df_candidatos_completo['regiao'] = df_candidatos_completo['estado'].map(estado_regiao)

media_idade_regiao = df_candidatos_completo.groupby('regiao')['idade'].mean()

plt.figure(figsize=(8, 6))
plt.bar(media_idade_regiao.index, media_idade_regiao.values, color='#3498db')
plt.title('Média de Idade por Região', fontsize=16)
plt.xlabel('Região', fontsize=14)
plt.ylabel('Média de Idade', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

for i, v in enumerate(media_idade_regiao.values):
    plt.annotate(f"{v:.1f}", (i, v), textcoords="offset points", xytext=(0, 5), ha='center')

plt.show()


print(media_idade_regiao)


```

# Analise 10

temos duas tabelas que mostam a quantidade de pessoas na area de atuaçao e o segundo grafico mostra a categoria de cada area que foi dividido em humanas, exatas e ciencias

```python
atuacao = df_candidatos_completo['area_atuacao'].value_counts()
humanas = ['Recursos Humanos', 'Educação', 'Administração', 'Marketing']
exatas = ['Tecnologia', 'Engenharia', 'Financeiro']
ciencias = ['Saúde']

def categorizar_area(area):
    if pd.isnull(area):
        return 'Não informado'
    elif area in humanas:
        return 'Humanas'
    elif area in exatas:
        return 'Exatas'
    elif area in ciencias:
        return 'Ciências Biológicas'
    else:
        return 'Outras'

df_candidatos_completo['categoria_area'] = df_candidatos_completo['area_atuacao'].apply(categorizar_area)
atuacao_categoria = df_candidatos_completo['categoria_area'].value_counts()

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
atuacao = df_candidatos_completo['area_atuacao'].value_counts(dropna=False)
plt.bar(atuacao.index, atuacao.values)
for i, v in enumerate(atuacao.values):
    plt.text(i, v + 1, f"{v}", ha='center')
plt.title('Distribuição das Áreas de Atuação', fontsize=16)
plt.xlabel('Área de Atuação', fontsize=14)
plt.ylabel('Quantidade', fontsize=14)
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
plt.bar(atuacao_categoria.index, atuacao_categoria.values)
for i, v in enumerate(atuacao_categoria.values):
    plt.text(i, v + 1, f"{(v / atuacao_categoria.sum() * 100):.1f}% ({v})", ha='center')
plt.title('Distribuição das Áreas de Atuação por Categoria', fontsize=16)
plt.xlabel('Categoria', fontsize=14)
plt.ylabel('Quantidade', fontsize=14)

plt.tight_layout()
plt.show()
```

# Analise 11


aqui temos uma analise que temos 624 vagas:
dessas vagas 46,63% o salario e maior que a pretensao dos candidatos;
47,60 dessas vagas os candidatos queriam ganhar mais que a vaga oferece;
5,77 estao com o salario satisfeito com o que e oferecido

```python
df_candidatos_completo = df_candidatos_completo.copy()
df_candidatos_completo['diferenca_salarial'] = df_candidatos_completo['salario_da_vaga'] - df_candidatos_completo['pretensoes_salariais']

qnt_vagas = df_candidatos_completo.shape[0]
media_diferenca_salarial = df_candidatos_completo['diferenca_salarial'].mean()
maior_que_pretensao = (df_candidatos_completo['diferenca_salarial'] > 0).sum() / qnt_vagas * 100
menor_que_pretensao = (df_candidatos_completo['diferenca_salarial'] < 0).sum() / qnt_vagas * 100
igual_pretensao = (df_candidatos_completo['diferenca_salarial'] == 0).sum() / qnt_vagas * 100

fig, axs = plt.subplots(1, 1, figsize=(8, 6))
barras = axs.bar(['Maior', 'Menor', 'Igual'], [maior_que_pretensao, menor_que_pretensao, igual_pretensao], color=['#4CAF50', '#FF9800', '#2196F3'])
axs.set_title('Porcentagem de Vagas por Diferença Salarial')
axs.set_xlabel('Diferença Salarial')
axs.set_ylabel('Porcentagem')
for i, v in enumerate([maior_que_pretensao, menor_que_pretensao, igual_pretensao]):
    axs.text(i, v + 1, f"{v:.2f}%", ha='center')
axs.legend(barras, ['Salário maior que a pretensão', 'Salário menor que a pretensão', 'Salário igual à pretensão'], loc='upper right')

plt.tight_layout()
plt.show()

print(f"Quantidade de vagas: {qnt_vagas}")
print(f"Média da diferença salarial: {media_diferenca_salarial:.2f}")
print(f"Porcentagem de vagas com salário maior que a pretensão salarial: {maior_que_pretensao:.2f}%")
print(f"Porcentagem de vagas com salário menor que a pretensão salarial: {menor_que_pretensao:.2f}%")
print(f"Porcentagem de vagas com salário igual à pretensão salarial: {igual_pretensao:.2f}%")
```
