```python
df_candidatos_completo = pd.concat([df_candidatos_tecnologia1, df_candidatos_tecnologia2, df_candidatos_tecnologia3], ignore_index=True)
df_candidatos_completo
```

```python
formacao = df_candidatos_completo['formacao'].value_counts()

plt.bar(formacao.index, formacao.values)
plt.show()
```

```python
# Ensure formacao_mapping is defined
formacao_mapping = {'Ensino Médio': 1, 'Tecnólogo': 2, 'Graduação': 3, 'Mestrado': 4, 'Doutorado': 5}

df_candidatos_completo['formacao_numerico'] = df_candidatos_completo['formacao'].map(formacao_mapping)

df_candidatos_completo['porcentagem'] = (df_candidatos_completo['formacao_numerico'] / df_candidatos_completo['formacao_numerico'].sum()) * 100
```

```python
df_candidatos_completo['formacao'] = df_candidatos_completo['formacao'].fillna('Unknown')

plt.bar(df_candidatos_completo['formacao'], df_candidatos_completo['porcentagem'])
plt.xlabel('formacao')
plt.ylabel('porcentagem')
plt.title('porcentagem por categoria')
plt.show()
```
