# Metodo One hot Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()

for serie in ['PAIS', 'TRANSMISSAO', 'MODELO']:
    df[serie] = lb.fit_transform(df[serie])
```

```python
df_dados = pd.read_excel('dados.xlsx')

legendas = {}

for var in ['PAIS', 'TRANSMISSAO', 'MODELO']:
    lb = LabelEncoder()  

    df_originais = df_dados[var].copy()

    # Fit and transform the data
    df_dados[var] = lb.fit_transform(df_originais)

    # Create a legend mapping
    legenda_var = dict(zip(lb.classes_, lb.transform(lb.classes_)))
    legendas[var] = legenda_var

# Print the legends
for var, mapa in legendas.items():
    print(f"\nLegenda para {var}:")
    for k, v in mapa.items():
        print(f" {k} : {v}")
```

```python
paises = ['Argentina', 'Brasil', 'Chile']

lb2 = LabelEncoder()

lb2.fit_transform(paises)

print(lb2.classes_)
```

```python
print(lb2.transform(lb2.classes_))
```

```python
print(dict(zip(lb2.classes_,lb2.transform(lb2.classes_))))
```
