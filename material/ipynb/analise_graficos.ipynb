# üìä An√°lise Gr√°fica de Dados com Pandas, Seaborn e Matplotlib

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar estilo dos gr√°ficos
sns.set(style="whitegrid")

# Carregar a base de dados
df = pd.read_csv('base_dados_graficos.csv')
df.head()
```

## üìä Histograma ‚Äî Distribui√ß√£o de Idades

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Idade"], bins=20, kde=False, color="skyblue")
plt.title("Distribui√ß√£o de Idades")
plt.xlabel("Idade")
plt.ylabel("Frequ√™ncia")
plt.show()
```

## üåä Gr√°fico de Densidade (KDE) ‚Äî Renda Mensal

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df["Renda Mensal (R$)"], fill=True, color="green")
plt.title("Densidade de Renda Mensal")
plt.xlabel("Renda (R$)")
plt.ylabel("Densidade")
plt.show()
```

## üì¶ Boxplot ‚Äî Renda por G√™nero

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="G√™nero", y="Renda Mensal (R$)", data=df, palette="pastel")
plt.title("Boxplot da Renda por G√™nero")
plt.xlabel("G√™nero")
plt.ylabel("Renda Mensal (R$)")
plt.show()
```

## üìã Gr√°fico de Barras ‚Äî Categorias Profissionais

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="Categoria Profissional", data=df, palette="muted",
              order=df["Categoria Profissional"].value_counts().index)
plt.title("Distribui√ß√£o por Categoria Profissional")
plt.xlabel("Categoria Profissional")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```

## ü•ß Gr√°fico de Pizza ‚Äî G√™nero

```python
df["G√™nero"].value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=["#ff9999", "#66b3ff"],
    figsize=(6, 6),
    startangle=90,
    wedgeprops=dict(width=0.5)
)
plt.title("Distribui√ß√£o de G√™nero")
plt.ylabel("")
plt.show()
```

## üü° Gr√°fico de Dispers√£o ‚Äî Idade vs. Satisfa√ß√£o com a Vida

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Idade", y="Satisfa√ß√£o com a Vida (0-10)", hue="G√™nero", data=df, alpha=0.5)
plt.title("Idade vs. Satisfa√ß√£o com a Vida")
plt.xlabel("Idade")
plt.ylabel("Satisfa√ß√£o com a Vida (0-10)")
plt.show()
```
