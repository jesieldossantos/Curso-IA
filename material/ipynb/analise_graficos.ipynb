# ðŸ“Š AnÃ¡lise GrÃ¡fica de Dados com Pandas, Seaborn e Matplotlib

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar estilo dos grÃ¡ficos
sns.set(style="whitegrid")

# Carregar a base de dados
df = pd.read_csv('base_dados_graficos.csv')
df.head()
```

## ðŸ“Š Histograma â€” DistribuiÃ§Ã£o de Idades

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Idade"], bins=20, kde=False, color="skyblue")
plt.title("DistribuiÃ§Ã£o de Idades")
plt.xlabel("Idade")
plt.ylabel("FrequÃªncia")
plt.show()
```

## ðŸŒŠ GrÃ¡fico de Densidade (KDE) â€” Renda Mensal

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df["Renda Mensal (R$)"], fill=True, color="green")
plt.title("Densidade de Renda Mensal")
plt.xlabel("Renda (R$)")
plt.ylabel("Densidade")
plt.show()
```

## ðŸ“¦ Boxplot â€” Renda por GÃªnero

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="GÃªnero", y="Renda Mensal (R$)", data=df, palette="pastel")
plt.title("Boxplot da Renda por GÃªnero")
plt.xlabel("GÃªnero")
plt.ylabel("Renda Mensal (R$)")
plt.show()
```

## ðŸ“‹ GrÃ¡fico de Barras â€” Categorias Profissionais

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="Categoria Profissional", data=df, palette="muted",
              order=df["Categoria Profissional"].value_counts().index)
plt.title("DistribuiÃ§Ã£o por Categoria Profissional")
plt.xlabel("Categoria Profissional")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```

## ðŸ¥§ GrÃ¡fico de Pizza â€” GÃªnero

```python
df["GÃªnero"].value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=["#ff9999", "#66b3ff"],
    figsize=(6, 6),
    startangle=90,
    wedgeprops=dict(width=0.5)
)
plt.title("DistribuiÃ§Ã£o de GÃªnero")
plt.ylabel("")
plt.show()
```

## ðŸŸ¡ GrÃ¡fico de DispersÃ£o â€” Idade vs. SatisfaÃ§Ã£o com a Vida

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Idade", y="SatisfaÃ§Ã£o com a Vida (0-10)", hue="GÃªnero", data=df, alpha=0.5)
plt.title("Idade vs. SatisfaÃ§Ã£o com a Vida")
plt.xlabel("Idade")
plt.ylabel("SatisfaÃ§Ã£o com a Vida (0-10)")
plt.show()
```
