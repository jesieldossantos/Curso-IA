```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 4, 3]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, label = 'informaçoes')
plt.ylabel('eixo y')
plt.xlabel('eixo x')
plt.title('Título do Gráfico')
plt.title('grafico de linha')
plt.xticks([2, 4, 6, 8, 10])
plt.yticks([2, 3, 4, 5, 6])
plt.legend()
plt.show()
```

```python
plt.plot(x, y, ls='--', lw=10, color='red')
plt.show()
```

```python
#Graficos de barras

plt.bar(x, y)
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('Título do Gráfico')
plt.title('grafico de barras')
plt.xticks([2, 4, 6, 8, 10])
plt.yticks([2, 3, 4, 5, 6])
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, s=100, c='red', alpha=0.5)
plt.show()
```

```python
figura = plt.figure(figsize=(20, 3))
figura.add_subplot()
plt.plot(x, y)
plt.title('grafico de linha')

figura.add_subplot()
plt.bar(x, y)
plt.title('grafico de barras')
```
