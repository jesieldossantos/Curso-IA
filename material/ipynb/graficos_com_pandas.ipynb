# Aula: Gráficos com Pandas - Estatística Visual
    
Nesta aula, vamos explorar como criar gráficos diretamente com a biblioteca Pandas utilizando um conjunto de dados fictício de alunos.
    

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura do dataset
df = pd.read_csv('dados_alunos.csv')
df.head()
```

```python
# Histograma de Idade
df['idade'].plot.hist(bins=10, title='Distribuição de Idades', figsize=(8, 4))
plt.xlabel('Idade')
plt.show()
```

```python
# Gráfico de Barras - Frequência de Gêneros
df['genero'].value_counts().plot.bar(title='Distribuição por Gênero', figsize=(6, 4))
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.show()
```

```python
# Gráfico de Barras - Quantidade de Alunos por Curso
df['curso'].value_counts().plot(kind='bar', title='Alunos por Curso', figsize=(8, 4))
plt.xlabel('Curso')
plt.ylabel('Total de Alunos')
plt.show()
```

```python
# Gráfico de Linhas - Nota Final Média por Semestre
df.groupby('semestre')['nota_final'].mean().plot(title='Nota Média por Semestre', marker='o')
plt.ylabel('Nota Final Média')
plt.xlabel('Semestre')
plt.grid(True)
plt.show()
```

```python
# Gráfico de Dispersão - Idade vs Nota Final
df.plot.scatter(x='idade', y='nota_final', title='Idade x Nota Final', alpha=0.7)
plt.show()
```

```python
# Boxplot - Comparação de Idade, Nota e Faltas
df[['idade', 'nota_final', 'faltas']].plot.box(title='Boxplot de Idade, Nota Final e Faltas')
plt.show()
```
