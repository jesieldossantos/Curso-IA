```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
df_original = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', sep=",")
```

```python
df_original.columns = ['CLASS','ALCOHOL','MALICACID','ASH','ASHALCALINITY','MAGNESIUM','TOTALPHENOLS','FLAVONOIDS','NONFLAVONOIDSPHENOLS','PRONTHOCYANINS','COLORINTENSITY','HUE','OD280_OD315','PRONLINE']
```

```python
df_original.head()
```

```python
df_original.describe()
```

```python
plt.figure(figsize=(10, 5))
ax = sns.boxplot(data=df_original)
plt.show()
```

```python
obj_padronizacao = StandardScaler()
```

```python
df_padronizado = obj_padronizacao.fit_transform(df_original)
```

```python
df_padronizado
```

```python
df_padronizado = pd.DataFrame(df_padronizado, columns=df_original.columns)

df_padronizado.head()
```

```python
df_padronizado.describe()
```

```python
plt.figure(figsize=(10, 5))
ax = sns.boxplot(data=df_padronizado)
plt.show()
```

```python
obj_normaizacao = MinMaxScaler()
```

```python
df_normalizacao = obj_normaizacao.fit_transform(df_original)
```

```python
df_normalizado = pd.DataFrame(df_normalizacao, columns=df_original.columns)

df_normalizacao.head()
```

```python
df_normalizado.describe()
```

```python
plt.figure(figsize=(10, 5))
ax = sns.boxplot(data=df_normalizado)
plt.show()
```

# Algoritimo SVM

```python
VAR_PREDITORAS_ORIG = df_original.drop('CLASS', axis= 1)
```

```python
VAR_TARGET = df_original['CLASS']
```

```python
VAR_TARGET = df_original['CLASS']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(VAR_PREDITORAS_ORIG, VAR_TARGET, test_size=0.3, random_state=10)
```

```python
svm = svm.SVC(kernel= 'linear')
svm.fit(X_train, Y_train)
```

```python
resultados= svm.predict(X_test)
```

```python
score = accuracy_score(Y_test, resultados)

score
```
